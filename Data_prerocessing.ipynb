```python
#imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
```

```python
df_credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
df_credits.head()
```

```python
df_movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df_movies.head()
```

```python
# df_movies.release_date.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(15,16))
# plt.show()
```

```python
movies = df_movies.merge(df_credits,on='movie_id')
```

```python
movies.columns
```

```python
movies.columns
```

```python
movies = movies.drop(columns=['title_x', 'tagline', 'status', 'homepage', 
                                        'keywords','crew','vote_count',
                                       'tagline', 'spoken_languages', 'runtime',
                                       'popularity', 'production_companies', 'budget',
                                       'production_countries', 'revenue',
                                        'title_y', 'original_language'])

```

```python
movies.head()
```

```python
features = ['cast', 'genres']
for feature in features:
    movies[feature] = movies[feature].apply(literal_eval)
```

```python
features
```

```python
def get_list(meta_data):
    if isinstance(meta_data, list):
        names = [col['name'] for col in meta_data]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []

features = ['cast','genres']
for feature in features:
    movies[feature] = movies[feature].apply(get_list)
```

```python
movies.isna().sum()
```

```python
movies.isnull().sum()
```

```python
movies['overview'] = movies['overview'].replace(np.nan, 'unkown')
```

```python
hindi = pd.read_csv('bollywood_full_1950-2019.csv')
```

```python
hindi.head()
```

```python
hindi = hindi.drop(columns=['title_x', 'tagline', 'poster_path', 
                                        'wiki_link','title_y','is_adult', 'year_of_release',
                                       'tagline',  'runtime',
                                       'imdb_votes', 'story',
                                       'wins_nominations'])

hindi.rename(columns = {'actors':'cast', 'summary':'overview', 'imdb_id':'movie_id', 'imdb_rating':'vote_average'}, inplace = True)

hindi['genres'] = hindi['genres'].str.split('|')

hindi['cast'] = hindi['cast'].str.split('|')



```

```python
hindi.columns
```

```python
movies.columns
```

```python
def clean_data(actors):
    if isinstance(actors, list):
        return actors[:3]

features = ['cast']
for feature in features:
    hindi[feature] = hindi[feature].apply(clean_data)

hindi.head()
```

```python
hindi.isnull().sum()
```

```python
hindi['overview'] = hindi['overview'].replace(np.nan, 'unkown')
hindi['cast'] = hindi['cast'].replace(np.nan, 'unkown')
hindi['vote_average'] = hindi['vote_average'].replace(np.nan, 'unkown')
```

```python
movies.columns
```

```python
hindi.columns
```

```python
#### Combining the bollywood and hollywoord movies

total_movies = movies.append(hindi, ignore_index=True, sort=True)
```

```python
total_movies.head()
```

```python
total_movies.tail()
```

```python
total_movies.to_csv('movie_data.csv', index = False)
```
